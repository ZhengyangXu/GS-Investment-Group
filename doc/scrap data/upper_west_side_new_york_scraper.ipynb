```python
import http.client 
import json
import requests
import pandas as pd
```

```python
upper_west_zipcode = [10023,10024,10025]
```

```python
def get_info(zipcode):
    
    conn = http.client.HTTPSConnection("search.onboard-apis.com") 

    headers = { 
    'accept': "application/json", 
    'apikey': "2eaa5798188384665ba0ec1257955fce"} 
    
    conn.request("GET", "/propertyapi/v1.0.0/property/address?"+"postalcode="+str(zipcode)+"&page=1&pagesize=100",headers=headers) 

    res = conn.getresponse() 
    data = res.read() 

    # print(data.decode("utf-8"))
    
    content_json = json.loads(data)
    

    df = pd.DataFrame(content_json['property'])
    
    return df
```

```python
address_all = []
for zipcode in upper_west_zipcode:
    address_all.append(get_info(zipcode))
```

```python
addressall = pd.concat(address_all)
```

```python
def get_property_detail(id):
    
    conn = http.client.HTTPSConnection("search.onboard-apis.com") 

    headers = { 
    'accept': "application/json", 
    'apikey': "2eaa5798188384665ba0ec1257955fce"} 

    conn.request("GET", "/propertyapi/v1.0.0/property/detail?id="+str(id),headers=headers) 

    res = conn.getresponse() 
    data = res.read() 

    # print(data.decode("utf-8"))

    content_json = json.loads(data)


    df = pd.DataFrame(content_json['property'])

    return df
```
